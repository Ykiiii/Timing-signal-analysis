# 读取预处理

```python
import pandas as pd

# 指定CSV文件的路径
csv_file_path = 'data\electricity.csv'

# 使用read_csv()函数读取CSV文件
df = pd.read_csv(csv_file_path)

# 显示DataFrame的前几行
print(df.iloc[:2000,:])
```

```python
dfc = df.iloc[:2000,-1]
dfc.to_csv('data\electricity_cleaned.txt', header=False, index=False)
```

```python
import numpy as np
y = np.array(dfc)
x=14
y_out = np.concatenate((y, np.zeros(x-2000%x)), axis=0)
y_out = y_out.reshape((2000//x+1,x))
dfc_out = pd.DataFrame(y_out)
print(y_out.shape)
dfc_out.to_csv('data\electricity_cleaned.csv', header=False, index=False)
```
